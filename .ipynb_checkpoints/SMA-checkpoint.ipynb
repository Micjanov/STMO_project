## A Study of Slime Molds
### An Overview
Slime Molds are a subset of **Amoebozoa** (protists) that fall under the infraphylum **Mycetozoa**. This classification includes three distinct classes: **Myxogastria, Dictyosteliida, and Protosteloids**, although there is still ongoing debate regarding this categorization. Our main focus in this study is a species from the Myxogastria group, but we will briefly touch upon the other classes due to their unique attributes.

The Protosteloids group consists of amoebae that can generate fruiting bodies and spores. On the other hand, species from the Dictyosteliida group, often referred to as **cellular** slime molds, show a particular life cycle. When they have an ample food supply, they behave like typical unicellular amoebae. However, when food becomes scarce, they aggregate to form a **pseudoplasmodium** or **grex**, which has a significantly increased ability to migrate. When conditions are suitable, this grex can further differentiate into a **sorocarp** or fruiting body for spore dispersion. This complex life cycle challenges the traditional understanding of unicellularity, multicellularity, and individuality.

## Examination of the Slime Mold Algorithm
In this study, we delve into the subject of a slime mold algorithm, referred to as the **SMA**. The utilization of **Physarum polycephalum** for problem-solving has already yielded beneficial results. The SMA seeks to embody some distinctive attributes of P. polycephalum, aiming to leverage the organism's aptitude for problem-solving in addressing a broader range of issues.

### Biological Inspirations for Algorithms
As explored in the course content, specific techniques like closed-form solutions, gradient descent algorithms, deterministic algorithms, and others are efficient and practical in addressing a certain class of problems, predominantly linear ones with a defined solution space. However, when faced with intricate non-linear real-world tasks where the solutions space and gradient data are unidentified, these algorithms may falter. Deriving a deterministic solution to these problems is often unattainable or necessitates impractical amounts of computation, as seen in the graph coloring problem from the course. The sole approach to definitively ascertain the global optimum is by examining all possible combinations, which becomes infeasible for complicated problems. As a result, metaheuristic algorithms are commonly employed instead. While these algorithms can't guarantee the discovery of the global optimum, they can swiftly identify acceptable solutions. They often incorporate stochastic elements that inject randomness, aiding in the evasion of local optima and boosting the probability of identifying the global optimum. The source of inspiration for such algorithms is frequently found in nature.

Natural organisms have been solving complex problems for millions of years, independent of deterministic algorithms or elaborate mathematics. These solutions in nature can occasionally be applied directly, as seen in biomimicry. Prominent examples include high-speed trains modeled after the beak of kingfishers and velcro inspired by seed dispersion mechanisms in plants like Arctium lappa, a common species in Belgium.

Similarly, certain algorithms are crafted by emulating the behavior of natural species. These typically employ swarm-based methods where multiple individuals interact with their surroundings and other members in the swarm. These techniques are often inspired by social insects. Some examples include:

- Particle Swarm Optimization
- Fruit Fly Optimization
- Ant Colony Optimization
- Artificial Bee Colony

The SMA falls under this category of algorithms. Interestingly, despite Physarum polycephalum being a single cell (often of significant size), a swarm-based technique is employed to replicate its behavior. This seems natural, considering the slime mold does not possess a centralized brain. Its intelligence emerges in a similar fashion to the swarm intelligence of social insect species. The different "individuals" in the "population" can be seen as separate parts of the same slime mold.

(For completeness, it's worth mentioning that the natural concept of evolution also serves as inspiration for genetic algorithms.)

### The Mathematical Foundation
The algorithm we will delve into in this notebook was devised by Li and colleagues in 2020. (Citation: Li et al. 2020. Slime mould algorithm: A new method for stochastic optimization. Future Generation Computer Systems. 10.1016/j.future.2020.03.055) The authors implemented this algorithm in Python, Java, and MATLAB.

The primary concepts of the SMA are illustrated in the following figure, which originates from the primary article:

![SMA](https://upload.wikimedia.org/wikipedia/commons/1/13/Slime_mould_algorithm_logic.png)

The model displays three distinct behavioral modes: Wrapping Food, Approaching Food, and Finding Food. The mathematical formulations for these varying behaviors are delineated below.

**Finding Food:**

$\overrightarrow{X(t+1)} = rand(UB-LB) + LB$

Here, *UB* and *LB* denote the upper and lower limits for the search parameters, respectively. When this mode is activated, the individual is placed randomly within the solution space. This leads to a random sampling of the design space.

**Approaching Food:**

$\overrightarrow{X(t+1)} = \overrightarrow{X_b(t)} +\overrightarrow{vb}(W \times \overrightarrow{X_A(t)} -\overrightarrow{X_B(t)})$

In this case, $\overrightarrow{X_b}$ represents the location of the individual possessing the best fitness score, while $\overrightarrow{X_A}$ and $\overrightarrow{X_B}$ are randomly selected individuals from the population. W signifies the weight of the slime mold, and $\overrightarrow{vb}$ is an oscillating function (described further below). W is computed using the following equation:

$\overrightarrow{W}=
\begin{cases}
1 + r \times log\left(\dfrac{bF-S(i)}{bF-wF} +1\right) & \quad \text{Half of population with best Fitness}\\ 
1 - r \times log\left(\dfrac{bF-S(i)}{bF-wF} +1\right) & \quad \text{Other}
\end{cases}$

Here, *bF* and *wF* stand for the best and worst fitness of the current generation, *S(i)* denotes the fitness of the current individual, and *r* is an oscillating function (explained further below).

**Wrapping Food:**

$\overrightarrow{X(t+1)} = \overrightarrow{vc} \times \overrightarrow{X(t)}$

In this equation, $\overrightarrow{vc}$ refers to an oscillating function (detailed further below).

The choice of which behavioral mode to activate depends on two stochastic determinations. Initially, **$rand < z$**, where rand is a random value ranging from [0,1] and z is a predetermined variable, is employed to decide between the *Finding Food* mode and the other two modes. If this condition isn't fulfilled, $r < p$ is used to choose between the *Approaching Food* and *Wrapping Food* modes. Here, *r* is a value between [0,1], and *p* is calculated using the following formula:

$p = tanh|S(i) - DF|$

In this formula, S(i) stands for the fitness of the individual, and DF denotes the best fitness across all generations. The rationale behind this decision boundary is demonstrated in the figure below.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of values from -10 to 10 with step 0.1
x = np.arange(-10, 10, 0.1)

# Compute tanh(x)
y_tanh = np.tanh(x)

# Compute tanh(|x|)
y_tanh_abs = np.tanh(np.abs(x))

# Create a new figure
plt.figure()

# Plot tanh(x)
plt.plot(x, y_tanh, lw=2, label='tanh()')

# Plot tanh(|x|)
plt.plot(x, y_tanh_abs, lw=2, linestyle='dotted', label='tanh(abs())')

# Set labels
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

When the fitness of the individual is far from DF (best fitness over all generations) the factor *p* will quickly approach 1. Consequently, the chance of satisfying $r < p$ becomes increasingly large and the position of $\overrightarrow{X(t)}$ will be updated according to the position of $\overrightarrow{X_b}$ (*Approaching Food* mode)(and an oscillating impact of $\overrightarrow{X_A}$ and $\overrightarrow{X_B}$) more often than according to its own position (*Wrapping Food* mode).

This is further illustrated by the interactive figure below. This figure shows the chance of a certain behavioral mode being chosen according to the fitness of the individual relative to best fitness over all generations.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

def plot_function(d_z, d_lb, d_ub, d_s):
    x = np.linspace(d_lb, d_ub, 1000)
    y1 = np.ones_like(x)
    y2 = 1 - d_z / 1000 * np.ones_like(x)
    y3 = np.tanh(np.abs(x - d_s)) * (1 - d_z / 1000)
    
    plt.figure(figsize=(8, 5))
    plt.fill_between(x, y1, color='green', alpha=0.5, label='Finding Food')
    plt.fill_between(x, y2, color='lightblue', alpha=0.5, label='Wrapping Food')
    plt.fill_between(x, y3, color='orange', alpha=0.5, label='Approaching Food')
    plt.axvline(x=d_s, color='red', linewidth=3, label='DF')
    plt.xlabel('Fitness')
    plt.ylabel('Chance')
    plt.ylim(0, 1)
    plt.legend()
    plt.grid(True)
    plt.show()

widgets.interact(plot_function, d_z=(1, 100, 1), d_lb=(-100, -1, 1), d_ub=(1, 100, 1), d_s=(-100, 100, 1))
```

Below a similar graph is made but this time in function of the distance in one-dimensional space instead of the fitness. The fitness function itself is also shown.

```python
import numpy as np
import matplotlib.pyplot as plt

def fitness_2_multi(solution):
    """
    Multidimentional f8 function

    Inputs:

	- solution: The coordinate for an individual

    Outputs:

	- The fitness for these coordinates
    """
    return -solution * np.sin(np.sqrt(np.abs(solution)))


def plot_figures(d_lb, d_ub, d_s, d_z):
    """
    This function generates two plots based on the defined functions 
    and settings above.

    Args:
    d_lb (int): lower bound
    d_ub (int): upper bound
    d_s (float): best fitness
    d_z (int): a parameter for adjusting the filled area of the plot

    Returns:
    None
    """
    x = np.linspace(d_lb, d_ub, 1000)

    # First plot
    fig, (ax1, ax2) = plt.subplots(2, 1)
    ax1.fill_between(x, 1, color='green', label='Finding Food')
    ax1.fill_between(x, 1 - d_z / 1000, color='lightblue', label='Wrapping Food')
    ax1.fill_between(x, np.tanh(np.abs([fitness_2_multi(i) for i in x])-d_s)*(1-d_z/1000), color='orange', label='Approaching Food')
    ax1.set_xlim([d_lb, d_ub])
    ax1.set_ylim([0, 1])
    ax1.set_xlabel('Location')
    ax1.set_ylabel('Chance')
    ax1.set_title('Chance Of Behavior')
    ax1.legend()
    
    # Second plot
    ax2.plot(x, np.abs([fitness_2_multi(i) for i in x]))
    ax2.set_xlim([d_lb, d_ub])
    ax2.set_xlabel('Location')
    ax2.set_ylabel('Fitness')
    ax2.set_title('Fitness Function')

    plt.show()

# Set the parameters
d_lb = -200  # lower bound
d_ub = 150  # upper bound
d_s = fitness_2_multi(0)  # best fitness
d_z = 50  # a parameter for adjusting the filled area of the plot

# Generate the plots
plot_figures(d_lb, d_ub, d_s, d_z)
```

It can be seen that the chance of displaying *Wrapping Food* behavior is small until a local or global minimum is near. Each time the fitness function reaches near a minimum, individuals at this location are likely to display *Wrapping Food* behavior, where oscillating changes to the location are made to finetune the fit.

The full equation for the locations is as followed:

\begin{equation}
\overrightarrow{X(t+1)}=
\begin{cases}
rand(UB-LB) + LB & \quad \text{if } rand < z \\ 
\overrightarrow{X_b(t)} +\overrightarrow{vb}(W \times \overrightarrow{X_A(t)} -\overrightarrow{X_B(t)}) & \quad \text{if } r < p \\
\overrightarrow{vc} \times \overrightarrow{X(t)} & \quad \text{if } r \geq p
\end{cases}
\end{equation}

Slime molds are able to accomplish much of their complex behavior because of the oscillation of their membranes. Or more accurately, because of the changes to the oscillation based on their environment. SMA tries to replicate some of these advantages by using multiple oscillating functions. Specifically the functions $\overrightarrow{vb}$, $\overrightarrow{vc}$ and $W$. The equations for $\overrightarrow{vb}$ and $\overrightarrow{vc}$ are as followed:

$\overrightarrow{vb} = \left[-arctanh(-\left(\dfrac{t}{max_t}\right) + 1), arctanh(-\left(\dfrac{t}{max_t}\right) + 1)\right]$

$\overrightarrow{vc} = \left[-(1-\left(\dfrac{t}{max_t}\right)), 1-\left(\dfrac{t}{max_t}\right)\right]$

Where *t* is the current generation and *max_t* is the maximum amount of generations.

The figure below shows the behavior of these two functions when sampling a single point for each generation.

```python
# Prepare the data
x = np.arange(1, 1001)

# Generate random values following Uniform distribution
vb = np.array([np.random.uniform(-np.arctanh(-(s / 1001) + 1), np.arctanh(-(s / 1001) + 1)) for s in x])
vc = np.array([np.random.uniform(-(1 - (s / 1001)), (1 - (s / 1001))) for s in x])

# Plot the data
plt.figure()
plt.plot(x, vb, label='vb')
plt.plot(x, vc, label='vc')
plt.xlabel('Generation')
plt.ylabel('Variation')
plt.title('Oscillating functions')
plt.legend()

# Show the plot
plt.show()
```

From this figure, it is clear that the behavior is not only modified by the fitness but also by the current generation relative to the maximum amount of generations. In early generations, the behavior is more erratic with higher oscillations. This results in a wider exploration of the solution space. In later generations, the oscillation decreases, and the changes to the locations will be smaller and trend more towards the (local) optima.

*r* in the formula for $W$ is a random sampling for the following interval:

$r = \left[0, 1\right]$

Unlike the other oscillating mechanisms, this oscillation is not modulated by the generation. Instead, it is modified by the fitness of the current individual in the formula for W.

#### âš ï¸Modifications to the original methodâš ï¸
Upon scrutinizing the original paper, we discovered an intriguing point. The calculations laid out in the paper for identifying new positions appear to be employed directly, rather than being utilized as guides or directional cues for the individuals. A look at the original code from the authors corroborated our initial suspicion that the calculated results should indeed serve as the new positions. This presents a peculiar situation, and here's why.

- Initially, when invoking the *Wrapping Food* mode of behavior, the location is expected to oscillate in sync with *vc*. However, in subsequent generations, *vc* gradually approaches zero. This implies that the *wrapping food* equation doesn't oscillate around the initial point, but oscillates between the initial location and zero. This imparts a propensity towards zero to the algorithm. The visibility of this effect is dampened due to a large number of benchmark functions having their optimum value at zero. We conducted a test with the original implementation using a fitness function whose optimum was located differently. It was evident that a rogue population often gathers around zero, a point that isn't even a local optimum in this case. The original paper contains a graph showcasing the search history utilizing the same fitness function, albeit with two dimensions. This graph reveals a similar stray population around zero.

- Secondly, the *Approaching food* formula deploys the *W*eight (correlated to fitness) of the individual in the computation. However, the original location, for which the weight was established, is not employed. Instead, the fitness of the location is leveraged to modulate the distance from the location with the best fitness to the new location.

**Revisions:**
The first issue could be mitigated by altering the *Wrapping Food* equation to:

$\overrightarrow{X(t+1)} = \overrightarrow{X(t)} + \overrightarrow{vc} \times \overrightarrow{X(t)}$

Lastly, we adjusted the method so that every dimension of an individual utilizes the same behavioral mode, instead of assessing each condition separately for every dimension.

The original and the amended versions of the algorithm are visually compared first, followed by a simulation study to assess their performances. This will enable us to ascertain whether our modifications result in superior outcomes.

### Implementation SMA

We used the original source code of the authors (both their original algrotithm (OrinigalSMA) and their modified algortihm (BaseSMA)), added some documentation and implemented modifications where noted. Here, we modified the code so that the loop breaks after the fitness value stays constant after a number of epochs.

```python
from numpy import where, clip, logical_and, ones, array, ceil
from numpy.random import uniform
from copy import deepcopy

class Root:
    """
    This is the root of all implemented Slime Mold Algorithms. It serves as a base class providing necessary
    utility methods and attributes used in all child algorithm classes.
    """

    ID_MIN_PROB = 0  # Indicator of minimization problem
    ID_MAX_PROB = -1  # Indicator of maximization problem

    ID_POS = 0  # Indicator for Position
    ID_FIT = 1  # Indicator for Fitness

    EPSILON = 10E-10  # Small constant value to avoid division by zero

    def __init__(self, obj_func=None, lb=None, ub=None, problem_size=50, verbose=True):
        """
        Parameters
        ----------
        obj_func : function
            Objective function to be optimized
        lb : list
            Lower bounds of the search space
        ub : list
            Upper bounds of the search space
        problem_size : int, optional
            Number of decision variables
        verbose : bool, optional
            Whether to print progress of the algorithm
        """
        self.obj_func = obj_func

        # Initialization of lower and upper bounds
        if (lb is None) or (ub is None):
            if problem_size is None:
                print("Problem size must be an integer number")
                exit(0)
            elif problem_size <= 0:
                print("Problem size must be > 0")
                exit(0)
            else:
                self.problem_size = int(ceil(problem_size))
                self.lb = -1 * ones(problem_size)
                self.ub = 1 * ones(problem_size)
        else:
            # Validation of lower and upper bounds
            if isinstance(lb, list) and isinstance(ub, list) and not (problem_size is None):
                if (len(lb) == len(ub)) and (problem_size > 0):
                    if len(lb) == 1:
                        self.problem_size = problem_size
                        self.lb = lb[0] * ones(problem_size)
                        self.ub = ub[0] * ones(problem_size)
                    else:
                        self.problem_size = len(lb)
                        self.lb = array(lb)
                        self.ub = array(ub)
                else:
                    print("Lower bound and Upper bound need to be the same length. Problem size must be > 0")
                    exit(0)
            else:
                print("Lower bound and Upper bound need to be a list. Problem size is an integer number")
                exit(0)

        self.verbose = verbose
        self.epoch, self.pop_size = None, None  # Initialize epoch and population size to None
        self.solution, self.loss_train = None, []  # Initialize solution and training loss to None and empty list

    def create_solution(self, minmax=0):
        """
        Return the solution with 2 elements: position of solution and fitness of solution
        Parameters
        ----------
        minmax: int
            0 - minimization problem, else - maximization problem
        """
        position = uniform(self.lb, self.ub)  # Generate a random position within the bounds
        fitness = self.get_fitness_position(position=position, minmax=minmax)  # Get the fitness of the position
        return [position, fitness]

    def get_fitness_position(self, position=None, minmax=0):
        """
        Assumption that objective function always return the original value
        Parameters
        ----------
        position: 1-D numpy array
            Position in the search space
        minmax: int
            0 - minimization problem, 1 - maximization problem
        Returns
        -------
        Fitness of the given position
        """
        return self.obj_func(position) if minmax == 0 else 1.0 / (self.obj_func(position) + self.EPSILON)

    def get_sorted_pop_and_global_best_solution(self, pop=None, id_fit=None, id_best=None):
        """
        Sort population and return the sorted population and the best solution
        Parameters
        ----------
        pop: list
            Population list
        id_fit: int
            Identifier for fitness
        id_best: int
            Identifier for best solution
        Returns
        -------
        Sorted population list and best solution
        """
        sorted_pop = sorted(pop, key=lambda temp: temp[id_fit])  # Sort population based on fitness
        return sorted_pop, deepcopy(sorted_pop[id_best])  # Return sorted population and best solution

    def amend_position(self, position=None):
        """
        Clip position so that it remains within the defined bounds
        Parameters
        ----------
        position: numpy array
            Position to be clipped
        Returns
        -------
        Clipped position
        """
        return clip(position, self.lb, self.ub)  # Clip position with respect to lower and upper bounds

    def amend_position_random(self, position=None):
        """
        Amend the position randomly if it is out of bounds
        Parameters
        ----------
        position: numpy array
            Position to be amended
        Returns
        -------
        Amended position
        """
        return where(logical_and(self.lb <= position, position <= self.ub), position, uniform(self.lb, self.ub))

    def update_sorted_population_and_global_best_solution(self, pop=None, id_best=None, g_best=None):
        """
        Sort the population and update the current best solution.
        Return the sorted population and the new current best solution
        Parameters
        ----------
        pop: list
            Population list
        id_best: int
            Identifier for best solution
        g_best: list
            Current best solution
        Returns
        -------
        Sorted population list and updated best solution
        """
        sorted_pop = sorted(pop, key=lambda temp: temp[self.ID_FIT])  # Sort population based on fitness
        current_best = sorted_pop[id_best]  # Get the current best solution from sorted population
        # Update the global best solution if current best solution is better
        g_best = deepcopy(current_best) if current_best[self.ID_FIT] < g_best[self.ID_FIT] else deepcopy(g_best)
        return sorted_pop, g_best  # Return sorted population and updated global best solution
```

The original implementation.

```python
from numpy.random import uniform, choice
from numpy import abs, zeros, log10, where, arctanh, tanh

class OriginalSMA(Root):
    """
    Original version of Slime Mould Algorithm (SMA)
    """
    ID_WEI = 2  # Index of the 'weight' component of a solution in the population

    def __init__(self, obj_func=None, lb=None, ub=None, problem_size=50, verbose=True, epoch=750, pop_size=100, z=0.03, improvement_threshold=1e-10, max_stagnant_iterations=50):
        """
        Initialize OriginalSMA
        """
        # Initialize root with objective function, lower bound, upper bound, problem size, and verbosity level
        Root.__init__(self, obj_func, lb, ub, problem_size, verbose)
        self.epoch = epoch  # Number of training iterations
        self.pop_size = pop_size  # Population size
        self.z = z  # Parameter used in position update
        self.improvement_threshold = improvement_threshold  # Fitness improvement threshold
        self.max_stagnant_iterations = max_stagnant_iterations  # Maximum number of stagnant iterations


    def create_solution(self, minmax=0):
        """
        Create a new solution
        """
        pos = uniform(self.lb, self.ub)  # Generate a random position within the bounds
        fit = self.get_fitness_position(pos)  # Get fitness of the position
        weight = zeros(self.problem_size)  # Initialize weight vector to zero
        return [pos, fit, weight]  # Solution is a combination of position, fitness, and weight

    def train(self):
        """
        Train using SMA
        """
        # Create initial population
        pop = [self.create_solution() for _ in range(self.pop_size)]
        # Sort the population and get the global best solution
        pop, g_best = self.get_sorted_pop_and_global_best_solution(pop, self.ID_FIT, self.ID_MIN_PROB)  # Eq.(2.6)

        for epoch in range(self.epoch):
            # Store old best fitness for convergence check
            old_best_fit = g_best[self.ID_FIT]
            s = pop[0][self.ID_FIT] - pop[-1][self.ID_FIT] + self.EPSILON  # Compute 's' parameter

            # Calculate the fitness weight of each slime mould
            for i in range(0, self.pop_size):
                # Eq.(2.5)
                if i <= int(self.pop_size / 2):
                    # First half of population gets increased weight
                    pop[i][self.ID_WEI] = 1 + uniform(0, 1, self.problem_size) * log10((pop[0][self.ID_FIT] - pop[i][self.ID_FIT]) / s + 1)
                else:
                    # Second half of population gets decreased weight
                    pop[i][self.ID_WEI] = 1 - uniform(0, 1, self.problem_size) * log10((pop[0][self.ID_FIT] - pop[i][self.ID_FIT]) / s + 1)

            # Compute 'a' and 'b' parameters for position update
            a = arctanh(-((epoch + 1) / self.epoch) + 1)  # Eq.(2.4)
            b = 1 - (epoch + 1) / self.epoch

            # Update the Position of search agents
            for i in range(0, self.pop_size):
                if uniform() < self.z:  # Eq.(2.7)
                    # Generate a completely new random position within bounds
                    pop[i][self.ID_POS] = uniform(self.lb, self.ub)
                else:
                    # Update position based on current fitness and positions of other individuals
                    p = tanh(abs(pop[i][self.ID_FIT] - g_best[self.ID_FIT]))  # Eq.(2.2)
                    vb = uniform(-a, a, self.problem_size)  # Eq.(2.3)
                    vc = uniform(-b, b, self.problem_size)
                    # For each dimension of the problem
                    for j in range(0, self.problem_size):
                        # Two random individuals selected from population
                        id_a, id_b = choice(list(set(range(0, self.pop_size)) - {i}), 2, replace=False)
                        if uniform() < p:  # Eq.(2.1)
                            # Update position based on best position and positions of two other individuals
                            pop[i][self.ID_POS][j] = g_best[self.ID_POS][j] + vb[j] * (pop[i][self.ID_WEI][j] * pop[id_a][self.ID_POS][j] - pop[id_b][self.ID_POS][j])
                        else:
                            # Update position based on current position
                            pop[i][self.ID_POS][j] = vc[j] * pop[i][self.ID_POS][j]

            # Check the bounds and calculate the fitness for each individual after all position updates are done
            for i in range(0, self.pop_size):
                pos_new = self.amend_position(pop[i][self.ID_POS])  # Check bounds of the new position
                fit_new = self.get_fitness_position(pos_new)  # Calculate fitness of the new position
                pop[i][self.ID_POS] = pos_new  # Update the position in the population
                pop[i][self.ID_FIT] = fit_new  # Update the fitness in the population

            # Sort the population and update the global best solution
            pop, g_best = self.update_sorted_population_and_global_best_solution(pop, self.ID_MIN_PROB, g_best)
            self.loss_train.append(g_best[self.ID_FIT])  # Append loss of global best solution
            if self.verbose:
                print("> Epoch: {}, Best fit: {}".format(epoch + 1, g_best[self.ID_FIT]))

            fitness_improvement = abs(old_best_fit - g_best[self.ID_FIT])
            if fitness_improvement < self.improvement_threshold:
                stagnant_iterations += 1  # Increase stagnant counter
                if stagnant_iterations >= self.max_stagnant_iterations:
                    print(f"Convergence achieved at iteration {epoch + 1}. Training stopped.")
                    break  # Stop training if stagnant for too long
            else:
                stagnant_iterations = 0  # Reset stagnant counter if significant improvement
                
        self.solution = g_best  # Update the best solution found
        return g_best[self.ID_POS], g_best[self.ID_FIT], self.loss_train  # Return the position, fitness, and loss of the best solution
```

```python
class BaseSMA(Root):
    """
    Modified version of Slime Mould Algorithm (SMA)
    """
    ID_WEI = 2  # Index of the 'weight' component of a solution in the population

    def __init__(self, obj_func=None, lb=None, ub=None, problem_size=50, verbose=True, epoch=750, pop_size=100, z=0.03, improvement_threshold=1e-10, max_stagnant_iterations=50):
        """
        Initialize BaseSMA
        """
        # Initialize root with objective function, lower bound, upper bound, problem size, and verbosity level
        Root.__init__(self, obj_func, lb, ub, problem_size, verbose)
        self.epoch = epoch  # Number of training iterations
        self.pop_size = pop_size  # Population size
        self.z = z  # Parameter used in position update
        self.improvement_threshold = improvement_threshold  # Fitness improvement threshold
        self.max_stagnant_iterations = max_stagnant_iterations  # Maximum number of stagnant iterations


    def create_solution(self, minmax=0):
        """
        Create a new solution
        """
        pos = uniform(self.lb, self.ub)  # Generate a random position within the bounds
        fit = self.get_fitness_position(pos)  # Get fitness of the position
        weight = zeros(self.problem_size)  # Initialize weight vector to zero
        return [pos, fit, weight]  # Solution is a combination of position, fitness, and weight

    def train(self):
        """
        Train using SMA
        """
        # Create initial population
        pop = [self.create_solution() for _ in range(self.pop_size)]
        # Sort the population and get the global best solution
        pop, g_best = self.get_sorted_pop_and_global_best_solution(pop, self.ID_FIT, self.ID_MIN_PROB)  # Eq.(2.6)
        stagnant_iterations = 0  # Initialize stagnant iterations counter
        for epoch in range(self.epoch):
            # Store old best fitness for convergence check
            old_best_fit = g_best[self.ID_FIT]
            s = pop[0][self.ID_FIT] - pop[-1][self.ID_FIT] + self.EPSILON  # Compute 's' parameter

            # Calculate the fitness weight of each slime mould
            for i in range(0, self.pop_size):
                # Eq.(2.5)
                if i <= int(self.pop_size / 2):
                    # First half of population gets increased weight
                    pop[i][self.ID_WEI] = 1 + uniform(0, 1, self.problem_size) * log10((pop[0][self.ID_FIT] - pop[i][self.ID_FIT]) / s + 1)
                else:
                    # Second half of population gets decreased weight
                    pop[i][self.ID_WEI] = 1 - uniform(0, 1, self.problem_size) * log10((pop[0][self.ID_FIT] - pop[i][self.ID_FIT]) / s + 1)

            # Compute 'a' and 'b' parameters for position update
            a = arctanh(-((epoch + 1) / self.epoch) + 1)  # Eq.(2.4)
            b = 1 - (epoch + 1) / self.epoch

            # Update the Position of search agents
            for i in range(0, self.pop_size):
                if uniform() < self.z:  # Eq.(2.7)
                    # Generate a completely new random position within bounds
                    pos_new = uniform(self.lb, self.ub)
                else:
                    # Update position based on current fitness and positions of other individuals
                    p = tanh(abs(pop[i][self.ID_FIT] - g_best[self.ID_FIT]))  # Eq.(2.2)
                    vb = uniform(-a, a, self.problem_size)  # Eq.(2.3)
                    vc = uniform(-b, b, self.problem_size)
                    # Two positions randomly selected from population
                    id_a, id_b = choice(list(set(range(0, self.pop_size)) - {i}), 2, replace=False)
                    pos_1 = g_best[self.ID_POS] + vb * (pop[i][self.ID_WEI] * pop[id_a][self.ID_POS] - pop[id_b][self.ID_POS])
                    pos_2 = vc * pop[i][self.ID_POS]
                    pos_new = where(uniform(0, 1, self.problem_size) < p, pos_1, pos_2)

                # Check if new position is within bounds
                pos_new = self.amend_position(pos_new)
                # Calculate fitness of the new position
                fit_new = self.get_fitness_position(pos_new)
                # Update position and fitness of the individual in the population
                pop[i][self.ID_POS] = pos_new
                pop[i][self.ID_FIT] = fit_new

            # Sort population and update the global best solution
            pop, g_best = self.update_sorted_population_and_global_best_solution(pop, self.ID_MIN_PROB, g_best)
            self.loss_train.append(g_best[self.ID_FIT])  # Append loss of global best solution
            if self.verbose:
                print("> Epoch: {}, Best fit: {}".format(epoch + 1, g_best[self.ID_FIT]))
            
            # Check if fitness has improved significantly
            fitness_improvement = abs(old_best_fit - g_best[self.ID_FIT])
            if fitness_improvement < self.improvement_threshold:
                stagnant_iterations += 1  # Increase stagnant counter
                if stagnant_iterations >= self.max_stagnant_iterations:
                    print(f"Convergence achieved at epoch {epoch + 1}. Training stopped.")
                    break  # Stop training if stagnant for too long
            else:
                stagnant_iterations = 0  # Reset stagnant counter if significant improvement
                
        self.solution = g_best  # Update the best solution found
        return g_best[self.ID_POS], g_best[self.ID_FIT], self.loss_train  # Return the position, fitness, and loss of the best solution
```

```python
class BaseSMA_follow(Root):
    """
    Modified version of Slime Mould Algorithm (SMA)
    """
    ID_WEI = 2  # Index of the 'weight' component of a solution in the population

    def __init__(self, obj_func=None, lb=None, ub=None, problem_size=50, verbose=True, epoch=750, pop_size=100, z=0.03, improvement_threshold=1e-10, max_stagnant_iterations=50):
        """
        Initialize BaseSMA
        """
        # Initialize root with objective function, lower bound, upper bound, problem size, and verbosity level
        Root.__init__(self, obj_func, lb, ub, problem_size, verbose)
        self.epoch = epoch  # Number of training iterations
        self.pop_size = pop_size  # Population size
        self.z = z  # Parameter used in position update
        self.improvement_threshold = improvement_threshold  # Fitness improvement threshold
        self.max_stagnant_iterations = max_stagnant_iterations  # Maximum number of stagnant iterations


    def create_solution(self, minmax=0):
        """
        Create a new solution
        """
        pos = uniform(self.lb, self.ub)  # Generate a random position within the bounds
        fit = self.get_fitness_position(pos)  # Get fitness of the position
        weight = zeros(self.problem_size)  # Initialize weight vector to zero
        return [pos, fit, weight]  # Solution is a combination of position, fitness, and weight

    def train(self):
        """
        Train using SMA
        """
        # Create initial population
        pop = [self.create_solution() for _ in range(self.pop_size)]
        # Sort the population and get the global best solution
        pop, g_best = self.get_sorted_pop_and_global_best_solution(pop, self.ID_FIT, self.ID_MIN_PROB)

        # Create empty lists to store fitness and location history
        fitness_history = []
        location_history = []

        stagnant_iterations = 0
        for epoch in range(self.epoch):
            # Store old best fitness for convergence check
            old_best_fit = g_best[self.ID_FIT]
            s = pop[0][self.ID_FIT] - pop[-1][self.ID_FIT] + self.EPSILON

            # Calculate the fitness weight of each slime mould
            for i in range(0, self.pop_size):
                if i <= int(self.pop_size / 2):
                    pop[i][self.ID_WEI] = 1 + uniform(0, 1, self.problem_size) * log10((pop[0][self.ID_FIT] - pop[i][self.ID_FIT]) / s + 1)
                else:
                    pop[i][self.ID_WEI] = 1 - uniform(0, 1, self.problem_size) * log10((pop[0][self.ID_FIT] - pop[i][self.ID_FIT]) / s + 1)

            a = arctanh(-((epoch + 1) / self.epoch) + 1)
            b = 1 - (epoch + 1) / self.epoch

            # Update the Position of search agents
            for i in range(0, self.pop_size):
                if uniform() < self.z: 
                    pos_new = uniform(self.lb, self.ub)
                else:
                    p = tanh(abs(pop[i][self.ID_FIT] - g_best[self.ID_FIT]))
                    vb = uniform(-a, a, self.problem_size)
                    vc = uniform(-b, b, self.problem_size)
                    id_a, id_b = choice(list(set(range(0, self.pop_size)) - {i}), 2, replace=False)
                    pos_1 = g_best[self.ID_POS] + vb * (pop[i][self.ID_WEI] * pop[id_a][self.ID_POS] - pop[id_b][self.ID_POS])
                    pos_2 = vc * pop[i][self.ID_POS]
                    pos_new = where(uniform(0, 1, self.problem_size) < p, pos_1, pos_2)

                pos_new = self.amend_position(pos_new)
                fit_new = self.get_fitness_position(pos_new)
                pop[i][self.ID_POS] = pos_new
                pop[i][self.ID_FIT] = fit_new

            # Update fitness and location histories
            fitness_history.append(np.mean([individual[self.ID_FIT] for individual in pop]))
            location_history.extend([individual[self.ID_POS] for individual in pop])

            # Continue the update of g_best if fitness improves
            pop, g_best = self.update_sorted_population_and_global_best_solution(pop, self.ID_MIN_PROB, g_best)
            if g_best[self.ID_FIT] - old_best_fit == 0:
                stagnant_iterations += 1
            else:
                stagnant_iterations = 0
            
            if stagnant_iterations >= self.max_stagnant_iterations:
                print(f"Convergence achieved at epoch {epoch + 1}. Training stopped.")
                break  # Stop training if stagnant for too long

        return g_best[self.ID_POS], g_best[self.ID_FIT], fitness_history, location_history
```

### Tests
First, we will look at the behavior of the algorithm in a one-dimensional setting. Working in one dimension makes it easier to visualize the behavior and reduces the computation need, however, the algorithm itself does not have a limit for the dimensions. Later we will also see a two-dimensional example.

The following fitness functions were used to test the behavior.

**Ackley Function:**

$$f(x_0 \cdots x_n) = -20 exp(-0.2 \sqrt{\frac{1}{n} \sum_{i=1}^n x_i^2}) - exp(\frac{1}{n} \sum_{i=1}^n cos(2\pi x_i)) + 20 + e$$

With n the amount of dimensions. This function was only used for the 1D setting.

**$f_8$:**

$$f_8(x_0 \cdots x_n) = \sum_{i=1}^n ( âˆ’x_i \times sin(\sqrt{| x_i |}))$$

With n the mount of dimensions. This function was used for both the 1D and 2D settings. This function is called $f_8$ according to the name used in the original paper. This function was selected because it has many local minima and because the global minimum is not located at the origin. This latter characteristic is usefull to show the problem with the original implementation. 

Note that these two fitness functions are symmetric in the dimensions. This is not a requirement and different dimensions can be evaluated using different functions as long as the end procuct of the method is a single fitness value. 
#### 1D Ackley Function

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
from math import exp, sqrt, pi

def func_ackley(solution):
    a, b, c = 20, 0.2, 2 * pi
    d = len(solution)
    sum_1 = -a * exp(-b * sqrt(np.sum(np.array(solution)**2) / d))
    sum_2 = np.exp(np.sum(np.cos(c * np.array(solution)) / d))
    return sum_1 - sum_2 + a + np.exp(1)

sma = BaseSMA_follow(obj_func=func_ackley, lb=[-100], ub=[100], problem_size=50, epoch=20, z=0.03)
best_pos, best_fit, loss_train, location_history = sma.train()

fig, axs = plt.subplots(2, 2)

def animate(i):
    axs[0, 0].cla()
    axs[0, 0].plot(loss_train[:i+1], label='Loss')
    axs[0, 0].set_title('Fitness')
    axs[0, 0].set(xlabel='Generation', ylabel='Fitness')
    axs[0, 0].legend()

    axs[0, 1].cla()
    axs[0, 1].scatter(range(i+1), [func_ackley(pos) for pos in location_history[:i+1]])
    axs[0, 1].set_title('Locations in Gen: {}'.format(i+1))
    axs[0, 1].set(ylabel='Location')

    axs[1, 0].cla()
    axs[1, 0].plot(np.linspace(-100, 100, num=200), [func_ackley([s]) for s in np.linspace(-100, 100, num=200)])
    axs[1, 0].scatter(best_pos, func_ackley(best_pos))
    axs[1, 0].set_title('Fitness Function')
    axs[1, 0].set(xlabel='Location', ylabel='Fitness')

    axs[1, 1].cla()
    axs[1, 1].scatter(range(i+1), [func_ackley(pos) for pos in location_history[:i+1]])
    axs[1, 1].vlines(i, min([func_ackley(pos) for pos in location_history]), max([func_ackley(pos) for pos in location_history]), colors='red')
    axs[1, 1].set_title('All Locations')
    axs[1, 1].set(xlabel='Generation', ylabel='Location')

ani = FuncAnimation(fig, animate, frames=range(len(loss_train)), repeat=False)
plt.show()
```
